```python
x = np.linspace(0, 2*np.pi, 100)

fig = plt.figure()

lines = plt.plot([0],[0])
line = lines[0]

# line, = plt.plot([])

#other setup
plt.xlim(0, 2*np.pi)
plt.ylim(-1.1, 1.1)

def animate(frame):
    # update plot
    y = np.sin(x+2*np.pi * frame/100)
    line.set_data((x,y))

anim = FuncAnimation(fig, animate, frames=100, interval=20)
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```

```python
fig = plt.figure()

# Initialize plot with placeholder data
lines = plt.plot([0], [0], 'o')
line = lines[0]

# Other setup
plt.axis("scaled")
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

def animate(frame):
    # Update plot
    x = np.cos(frame / 100 * 2 * np.pi)
    y = np.sin(frame / 100 * 2 * np.pi)
    line.set_data([x], [y])
    # return line,

# Create animation
anim = FuncAnimation(fig, animate, frames=100, interval=20)

# Display animation
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```
